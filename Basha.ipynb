```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
basha=pd.read_csv('energy.csv')
print(basha)
srh=basha[basha['State']=="Andhra"]
rcb=basha[basha['State']=="Karnataka"]
# LINE CHART ANDHRA KARANATAKA
fgi1,diag1=plt.subplots(2)
#date, Consumption
diag1[0].plot(srh['Date'],srh['Consumption'],color='orange')
diag1[1].plot(rcb['Date'],rcb['Consumption'],color='red')

#bar chart Telengana Tripura
akhil=basha[basha['State']=='Telengana']
cm=basha[basha['State']=='Tripura']

fig,diag=plt.subplots(2)
diag[0].bar(akhil['Date'],akhil['Consumption'],color='orange')
diag[1].bar(cm['Date'],cm['Consumption'],color='red')

#pie chart
state1='Delhi'
state2='Odisha'
state3='UP'
state4='Punjab'
cse5=[state1,state2]

cse4=basha[basha['State']==state1]['Consumption'].sum()
cse3=basha[basha['State']==state2]['Consumption'].sum()
cse2=[cse3,cse4]

fig,diag=plt.subplots(2)
diag[0].pie(cse2,labels=cse5,autopct='%1.1f%%')

cse5=[state3,state4]
cse4=basha[basha['State']==state3]['Consumption'].sum()
cse3=basha[basha['State']==state4]['Consumption'].sum()
cse2=[cse3,cse4]
diag[1].pie(cse2,labels=cse5,autopct='%1.1f%%')




```
